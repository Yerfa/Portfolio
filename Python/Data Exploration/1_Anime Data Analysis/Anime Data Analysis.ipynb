<font size=6>Analisis Banyak Rating Anime</font>

```python
RagamRating = df[['Name','Rating']].dropna(axis=0, how='all')
RagamRating = RagamRating.dropna(axis=0, subset=['Rating'])
RagamRating['Rating'] = RagamRating['Rating'].astype(int)
RagamRating = RagamRating.groupby('Rating', as_index=False).count()
```

```python
RagamRating.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
plt.bar(RagamRating['Rating'], RagamRating['Name'], color='green')
plt.title('Banyak Nilai Setiap Rating', size=16, pad=15)
plt.xlabel('Nilai Rating',size=14)
plt.ylabel('Banyak Rating', size=14)
plt.show()
```

```python
RagamRating2 = df[['Name','Rating']].dropna(axis=0, how='all').dropna(axis=0, subset=['Rating'])
```

```python
RagamRating2.isna().sum()
```

```python
RagamRating2 = RagamRating2.loc[(RagamRating2['Rating'] >= 3) & (RagamRating2['Rating'] < 4)]
```

```python
RagamRating2['Rating'] = RagamRating2['Rating'].apply(lambda x: round(x, 1))
print(RagamRating2['Rating'].unique())
```

```python
RagamRating2 = RagamRating2.groupby('Rating', as_index=False)['Name'].count()
print(RagamRating2)
```

```python
RagamRating2.set_index('Rating', drop=True).plot(kind='bar')
plt.title('Grafik Rating 3 Sampai 4 dengan Index', size=16, pad=15)
plt.xlabel('Nilai Rating', size=14)
plt.ylabel('Banyak Rating', size=14)
plt.xticks(rotation=0)
plt.show()
```

<font size='6'>Presentasi Keamanan Konten Anime</font>

```python
ContentAbuse = df.copy()
ContentAbuse['Content_Warning'] = ContentAbuse['Content_Warning'].apply(lambda x: False if x is np.NaN else True)
```

```python
ContentAbuse['Content_Warning'].unique()
```

```python
PiePie = ContentAbuse.groupby('Content_Warning')['Name'].count().reset_index()
PiePie['Name'] = pd.to_numeric(PiePie['Name'], downcast='float')
plt.pie(PiePie['Name'], labels = PiePie['Content_Warning'], autopct='%1.2f%%')
plt.title('Persentase Anime dengan dan tanpa Content Warning', loc='center', color='white', size=16)
plt.legend(title = 'Content Warning')
plt.show()
```

```python
PiePie = ContentAbuse.groupby('Content_Warning')['Name'].count().reset_index()
PiePie['Name'] = pd.to_numeric(PiePie['Name'], downcast='float')
plt.pie(PiePie['Name'], labels = PiePie['Content_Warning'])
plt.title('Persentase Anime dengan dan tanpa Content Warning', loc='center', color='white', size=16)
plt.legend(title = 'Content Warning', loc='center left', bbox_to_anchor=(1, 0.1, 0.5, 1))
plt.gcf().gca().add_artist(plt.Circle( (0,0), 0.6, color='white'))
plt.text(-0.5, -0.1, '90.04% Anime\nTidak Aman !!!', size=12, color='black')
plt.show()
```

```python
data = PiePie['Name'].astype(int)

wedges, texts = plt.pie(data, wedgeprops=dict(width=0.5), startangle=-40)
bbox_props = dict(boxstyle="square,pad=0.3", fc="w", ec="k", lw=0.72)
kw = dict(arrowprops=dict(arrowstyle="-"),
          bbox=bbox_props, zorder=0, va="center")
recipe = PiePie['Content_Warning']


for i, p in enumerate(wedges):
    ang = (p.theta2 - p.theta1)/2. + p.theta1
    y = np.sin(np.deg2rad(ang))
    x = np.cos(np.deg2rad(ang))
    horizontalalignment = {-1: "right", 1: "left"}[int(np.sign(x))]
    connectionstyle = "angle,angleA=0,angleB={}".format(ang)
    kw["arrowprops"].update({"connectionstyle": connectionstyle})
    plt.annotate(round(100*(data[i]/data.sum()),2), xy=(x, y), xytext=(1.35*np.sign(x), 1.4*y),
                horizontalalignment=horizontalalignment, **kw)
plt.title('Persentase Anime dengan dan tanpa Content Warning', loc='center', color='white', size=16)
plt.gcf().gca().add_artist(plt.Circle( (0,0), 0.6, color='white'))
plt.text(-0.5, -0.1, '90.04% Anime\nTidak Aman !!!', size=12, color='black')
plt.show()
```

<font size='6'>Top 10 Studio Anime Terbanyak</font>

```python
StudioAnime = df.copy()
StudioAnime = StudioAnime.loc[~StudioAnime.Studio.isna()]
```

```python
StudioAnime['Studio'] = StudioAnime['Studio'].astype(str)
```

```python
StudioAnime.dtypes
```

```python
StudioAnime['Studio'] = StudioAnime['Studio'].str.slice(0,10)
StudioChart = StudioAnime[['Studio','Name']].groupby('Studio').count().sort_values('Name', ascending=False).head(10)
StudioChart = StudioChart.rename(columns={'Name':'Banyak Anime'})
```

```python
plt.figure()
StudioChart.plot(kind='bar')
plt.title('Studio Terbanyak Membuat Anime', fontsize=18, pad=20)
plt.xlabel('Studio', fontsize=16, labelpad=10)
plt.ylabel('Banyak Anime', fontsize=16, labelpad=10)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
